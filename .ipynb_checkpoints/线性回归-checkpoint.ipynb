```python
import pandas as pd
import matplotlib.pyplot as plt
columns=["mpg","cylinders","displacement","horsepower","weight","acceleration","model year","origin","car name"]
cars=pd.read_csv("d://python/sklearn/auto-mpg.data",delim_whitespace=True,names=columns)
print(cars.head(5))
```

```python
fig=plt.figure()
ax1=fig.add_subplot(2,1,1)
ax2=fig.add_subplot(2,1,2)
cars.plot("weight","mpg",kind='scatter',ax=ax1)
cars.plot("acceleration","mpg",kind='scatter',ax=ax2)
plt.show()
```

```python
import sklearn
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fir(cars[["weight"]],cars["mpg"])
```

```python
lr=LinearRegression(fit_intercept=True)
lr.fit(cars[["weight"]],cars["mpg"])
predictions=lr.predict(cars[["weight"]])
print(predictions[0:5])
print(cars["mpg"][0:5])
```

```python
plt.scatter(cars["weight"],cars["mpg"],c='red')
plt.scatter(cars["weight"],predictions,c='blue')
plt.show()
```

```python
mse=sklearn.metrics.mean_squared_error(cars["mpg"],predictions)
print(mse)
rmse=mse**(0.5)
print(rmse)
```
